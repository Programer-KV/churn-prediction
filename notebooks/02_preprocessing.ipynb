```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(df.shape)
```

```python
df.dtypes
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ",0)
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df['Churn'] = df['Churn'].map({ 'Yes':1, 'No' : 0 })
```

```python
df= df.drop('customerID',axis = 1)
```

```python
categorical_cols = df.select_dtypes(include='object').columns
print(categorical_cols)
print(f"Total categorical columns: {len(categorical_cols)}")
```

```python
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
    
```

```python
binary_cols = ['gender','Partner','Dependents','PhoneService','PaperlessBilling']
binary_map = {'Yes': 1, 'No': 0, 'Male': 1,'Female': 0}
for col in binary_cols:
    df[col] = df[col].map(binary_map)

simplify_cols = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                 'DeviceProtection', 'TechSupport', 
                 'StreamingTV', 'StreamingMovies']

simplify_maps = {'Yes':1,'No':0,'No internet service':0,'No phone service':0}

for col in simplify_cols:
    df[col] = df[col].map(simplify_maps)

ohe_cols = ['InternetService', 'Contract', 'PaymentMethod']
df = pd.get_dummies(df, columns=ohe_cols, drop_first=True)
```

```python
print(df.shape)
print(df.dtypes)
```

```python
df.head()
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
simplify_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                 'TechSupport', 'StreamingTV', 'StreamingMovies']
df[simplify_cols] = df[simplify_cols].astype(int)
```

```python
for col in simplify_cols:
    print(f"{col}: {df[col].isnull().sum()} nulls")
    
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)

df.to_csv('../data/processed/telco_processed.csv', index=False)
print("Saved!")
print(df.shape)
```
